# Introduccion

El juego Pokémon es uno de los títulos más icónicos y populares de la industria de los videojuegos, capturando la imaginación de millones de jugadores en todo el mundo desde su lanzamiento. Con una rica historia que abarca varias generaciones, Pokémon ha evolucionado continuamente, ofreciendo nuevas experiencias, regiones, y criaturas en cada entrega. A lo largo de los años, se han lanzado innumerables versiones y variantes del juego, cada una con su propia colección de Pokémon, algunos exclusivos de ciertas versiones, lo que incentiva el intercambio entre jugadores.

Uno de los objetivos principales de la serie es la captura de Pokémon salvajes, que luego pueden ser añadidos al equipo del jugador para participar en batallas estratégicas. La emoción de encontrar, atrapar y entrenar a estos Pokémon es el corazón de la experiencia de juego. Cada Pokémon tiene sus propias estadísticas, habilidades y movimientos, lo que hace que la elección de un equipo balanceado sea esencial para enfrentar con éxito los desafíos del juego.

Para formar un equipo efectivo, es crucial considerar varios factores: ¿Qué tan difícil es capturar ciertos Pokémon? ¿Cuáles son los más fuertes o los que poseen habilidades únicas que pueden dar una ventaja en combate? Estos aspectos son fundamentales para avanzar en el juego.

En este contexto, resulta útil emplear técnicas de análisis de datos en Python, como las que hemos aprendido en el curso, para manipular bases de datos y visualizar información relevante. Al hacerlo, podremos responder de manera efectiva a preguntas clave y optimizar nuestras estrategias dentro del juego.

```python
# importar las librerias que usaremos
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
```

```python
# Montamos el sistema de archivos de Google Drive
drive.mount('/content/drive')

# Definimos la ruta al archivo CSV
file_path = '/content/drive/MyDrive/Coder_DS_1/Pokemon Database.csv'

# Ponemos como input el archivo CSV en un DataFrame de pandas
df = pd.read_csv(file_path, delimiter=',')

df
```

```python
# Eliminamos las algunas columnas por nombre
df = df.drop(['Pokedex Number', 'Classification', 'Original Pokemon ID', 'Pokemon Height', 'Pokemon Weight', 'Primary Ability', 'Primary Ability Description', 'Secondary Ability', 'Secondary Ability Description', 'Hidden Ability', 'Hidden Ability Description', 'Special Event Ability', 'Special Event Ability Description', 'Male Ratio', 'Female Ratio', 'Base Happiness', 'Game(s) of Origin', 'Health EV', 'Attack EV', 'Defense EV', 'Special Attack EV', 'Special Defense EV', 'Speed EV', 'EV Yield Total', 'Experience Growth', 'Experience Growth Total', 'Primary Egg Group', 'Secondary Egg Group', 'Egg Cycle Count', 'Evolution Details'], axis=1)
df
```

```python
# Reemplazamos los valores NaN en 'Pre-Evolution Pokemon Id' por 'Transformation' si 'Alternate Form Name' no es NaN
df['Pre-Evolution Pokemon Id'] = df.apply(lambda row: 'Transformation' if pd.notna(row['Alternate Form Name']) and pd.isna(row['Pre-Evolution Pokemon Id']) else row['Pre-Evolution Pokemon Id'], axis=1)
df
```

```python
#Llenamos los espacios vacios en las columnas y reemplazamos por valores de texto, excepto para Alternate Form Name
df.loc[:, 'Legendary Type'] = df['Legendary Type'].fillna('Normal')
df.loc[:, 'Secondary Type'] = df['Secondary Type'].fillna('None')
df.loc[:, 'Pre-Evolution Pokemon Id'] = df['Pre-Evolution Pokemon Id'].fillna('Base')
df
```

```python
# Concatenamos el nombre del Pokemon y el nombre de su forma alterna para crear una nueva columna con el nombre completo
df.insert(1, 'Pokemon', df['Pokemon Name'].str.cat(df['Alternate Form Name'], sep=' ', na_rep=''))

# Eliminamos las columnas que utilizamos para la concatenacion para no tener columnas redundantes
df = df.drop(['Pokemon Name', 'Alternate Form Name'], axis=1)

df
```

```python
#Reemplazamos las "" en el nombre completo del pokemon para comodidad visual
df = df.replace('"', '', regex=True)

df
```

```python
#Tendremos que algunos pokemones estaban repetidos debido a que se consideraba las evoluciones de diferentes metodos (Ej: hacerlo de dia o noche pero no cambiando su estadisticas, variable que no consideraremos)
df = df.drop_duplicates(subset=['Pokemon Id'], keep='first')
df
```

```python
#Sacamos los .0 de los numeros en la columna de Pre Evolucion ya que es mejor visualmente
df.loc[:, 'Pre-Evolution Pokemon Id'] = df['Pre-Evolution Pokemon Id'].astype(str).str.replace('.0', '', regex=False)
df
```

```python
#Por ultimo chequeamos que no haya valores faltantes en nuestro dataframe
df.isna().sum()
```

# Hipotesis 1 (Legendarios)

Para comenzar, podríamos analizar la relación entre las estadísticas totales de los Pokémon y su clasificación como legendarios. Esto nos permitirá evaluar si realmente es beneficioso tener Pokémon legendarios en tu equipo y si su inclusión ofrece una ventaja significativa en combate.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Legendary Type', y='Base Stat Total', data=df)
plt.xlabel('Clasificaion Legendaria')
plt.ylabel('Estadisticas Totales')
plt.title('Legendario vs. No Legendario')
plt.show()
```

## Conclusión:

 Podemos ver del gráfico que los Pokémon legendarios tienden a tener unas estadisticas totales significativamente mayores que los no legendarios. Por lo que podemos asumir que son frecuentemente mas fuertes y convienen para el combate, esto al notar mediana más alta.

 Aun asi esto no saca que podemos notar una variabilidad mucho mas alta de los legendarios es decir, te toca un pokemon debil legendario sera bastanbte debil pero un fuerte podria ser uno de los mas fuertes que los no legendarios. Estro complementado con la exiistencia de outliers.  

Por lo tanto priorizar pokemons legendarios puede ser una efectivo pero tiene cierto riesgo de conseguir uno debil.  

#Hipotesis 2 (Dificultad de atrapar Pokemones fuertes)

Hemos visto que el estado legendario normalmente presenta una ventaja para nuestro equipo, pero no debemos olvidar que para conseguirlos hay que capturarlos.

A la hora de facilitar la completicion del juego no solo queremos tener un equipo fuerte sino tambien uno que no tome muchas mas horas de obtener que otro un poco menos fuerte pero mas facil de obtener.

En otras palabras, queremos ver cuales son los pokemones mas fuertes y tambien mas sencillos de capturar ademas de ver sus estados legendarios

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Base Stat Total', y='Catch Rate', hue='Legendary Type', data=df)
plt.xlabel('Estadisticas Totales')
plt.ylabel('Ratio de captura')
plt.title('Ratio de captura segun las estadistaicas totales y clasificacion')
plt.show()
```

## Conclusiones:

Podemos ver que los pokemons mas fuertes son en efecto mas dificiles de capturar.

Y no solo eso sino que los legendarios al ser los mas fuertes nomralemente, son mas dificiles de capturar.

Pero no todo esta perdido, vemos que algunos son mas sencillos de capturar y ademas que hay pokemones normales fuertes que son mas sencillos de capturar

Entonces nuestra estrategia constara de capturar los pokemones legendarios y normales fuertes sencillos de capturar, y no los mas poderosos y solamente legendarios.
